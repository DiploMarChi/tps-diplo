```python
#HPC BOT IPUCHA / SOLA PARDO

afiliado = input("Es usted afiliado? (Si/No): ").strip().capitalize()
finalizar_programa = False

while not finalizar_programa:
    if afiliado == "Si":
        print('Bienvenido')
        break
    elif afiliado == "No":
        lista_sector = [ # solo se puede elegir diagnostico por imagenes
            "Diagnóstico por Imágenes",
            "Laboratorio de Análisis",
            "Recetas Digitales",
            "Estudios de Audiología/Otorrinolaringología",
            "Estudios de Cardiología",
            "Otras consultas",
            "Afiliarse a Planes de Salud"
        ]

        print('Le pedimos que nos indique el sector con el cual desea comunicarse, indicando el número de opción correspondiente:')

        for i, opcion in enumerate(lista_sector, 1):
            print(f"{i}. {opcion}")

        try:
            eleccion = int(input("\nIngrese el número de la opción: "))
            if 1 <= eleccion <= len(lista_sector):
                print(f"Usted ha seleccionado: {lista_sector[eleccion - 1]}, ¿En qué podemos ayudarlo?")

                if eleccion == 1:  # solo se puede elegir radiografia
                    while True:
                        lista_Diagnostico_por_imagenes = [
                            "Realizar radiografía",
                            "Realizar PET",
                            "Turno otros estudios",
                            "Reprogramar mi turno",
                            "Cancelar mi turno",
                            "Retirar estudio",
                            "Menú anterior"
                        ]

                        print('Le pedimos que nos indique qué desea realizar, indicando el número de opción correspondiente:')
                        for i, opcion in enumerate(lista_Diagnostico_por_imagenes, 1):
                            print(f"{i}. {opcion}")

                        try:
                            eleccion2 = int(input("\nIngrese el número de la opción: "))
                            if 1 <= eleccion2 <= len(lista_Diagnostico_por_imagenes):
                                if eleccion2 == 7:
                                    break
                                else:
                                    print(f"Usted ha seleccionado: {lista_Diagnostico_por_imagenes[eleccion2 - 1]}")
                                    print("\nLe informamos que las placas simples o radiografías se realizan sin turno de \nlunes a viernes de 8 a 19:30 hs. en el Servicio de Diagnóstico \npor Imágenes (Córdoba N°4545). \nRecuerde que deberá presentar la orden médica original \n(si es digital del HPC no es necesario), \ncredencial de la cobertura y DNI. \nSaludos cordiales, HPC.")

                            lista_bucle = [
                                "Menú anterior",
                                "Menú principal",
                                "Finalizar"
                            ]

                            for i, opcion in enumerate(lista_bucle, 1):
                                print(f"{i}. {opcion}")

                            eleccion_bucle = int(input("\n¿Lo podemos ayudar en algo más? Ingrese el número de opción correspondiente: "))
                            if eleccion_bucle == 1:
                                continue
                            elif eleccion_bucle == 2:
                                break
                            elif eleccion_bucle == 3:
                                print("Gracias por su consulta. ¡Hasta luego!")
                                finalizar_programa = True
                                break
                            else:
                                print("Opción inválida, por favor elija un número válido.")
                        except ValueError:
                            print("Entrada no válida. Por favor, ingrese un número.")
                if finalizar_programa:
                    break
            else:
                print("Opción inválida, por favor elija un número válido.")
        except ValueError:
            print("Entrada no válida. Por favor, ingrese un número.")
    else:
        print("Por favor, responda con 'Si' o 'No'.")
        afiliado = input("Es usted afiliado? (Si/No): ").strip().capitalize()
```
