## 3 - Funciones de Busqueda ##

```python
def buscar_lista_precios_x_codigo(lista_precios, codigo):
  for item in lista_precios:
      if item["codigo"] == codigo:
        return item

def codigos_lista_precios(lista_precios):
  codigos = []
  for item in lista_precios:
      codigos.append(item["codigo"])
  return codigos
```

### 7.2 - Solicitar pedido al usuario ###

Funcion para obtener el pedido del cliente. Retorna una lista con el siguiente formato:<br>
[*nro_pedido, lista-pizzas,lista-empandas,lista-bebidas*]<br>
donde:



*   *nro_pedido*: es el numero de pedido registrado por el bot
*   *lista-pizzas*: [pedido-pizza-1, pedido-pizza-2, ...]
*   *lista-empanadas*: [pedido-empanada-1, pedido-empanada-2, ...]
*   *lista-bebidas*: [pedido-bebida-1, pedido-bebida-2, ...]

Ejemplo:
[1, [[1,'G',2],[2,'C',1]],[[1,'F',2],[3,'H',4]],[]]

Corresponde a ...<br>
Pedido #: 1<br>
Pizzas: 2 grandes de la variedad 2, 1 chica de la variedad 2<br>
Empanadas: 2 fritas de la variedad 1, 4 horno de la variedad 3<br>
Bebidas: ninguna





```python
def no_pidio_aun(pedido_pizzas, pedido_empanadas, pedido_bebidas):
  return pedido_pizzas == [] and pedido_empanadas == [] and pedido_bebidas == []


def tomar_pedido(nro_pedido):
  fin_pedido = False

  pedido_pizzas = []
  pedido_empanadas = []
  pedido_bebidas = []

  while not fin_pedido:

    listar_menu_ppal_variedades()

    if no_pidio_aun(pedido_pizzas, pedido_empanadas, pedido_bebidas):
      print('Pizzabot: ¿Qué queres llevar?')
    else:
      print('Pizzabot: ¿Algo mas que quieras llevar?')

    print()

    variedad_elegida = input('Vos: ')
    variedad_elegida = variedad_elegida.upper()

    if variedad_elegida.startswith('PIZZA'):
      pizzas = solicitar_pedido_pizzas()
      pedido_pizzas.append(pizzas)
    elif variedad_elegida.startswith('EMPANADA'):
      empanadas = solicitar_pedido_empanadas()
      pedido_empanadas.append(empanadas)
    elif variedad_elegida.startswith('BEBIDA'):
      bebidas = solicitar_pedido_bebidas()
      pedido_bebidas.append(bebidas)
    elif variedad_elegida == '':
      fin_pedido = True
    else:
      print('Pizzabot: No te entendi, intenta de nuevo ...')

  return [nro_pedido, pedido_pizzas, pedido_empanadas, pedido_bebidas]
```

### 8.2 - Muesta el detalle del pedido completo por pantalla ###

```python
def mostrar_detalle_pedido(pedido_cliente) -> None:

  print()
  print("Detalle Pedido:")
  print()
  print("Nro. Pedido # "+str(pedido_cliente["nro_pedido"]).zfill(3))
  print()

  # Imprimir detalle de pedidos de pizzas
  if pedido_cliente["pizzas"] != []:
    print("Pizzas:")
    lista_pizzas = pedido_cliente["pizzas"]
    for pizza in lista_pizzas:
      print(str(pizza[2]).rjust(2) + " " + pizza[0].ljust(25) + " " + pizza[1] + "."*(40-len(pizza[1])) + "$ " + str(pizza[3]).rjust(7))
    print()

  # Imprimir detalle de pedidos de empanadas
  if pedido_cliente["empanadas"] != []:
    print("Empanadas:")
    lista_empanadas = pedido_cliente["empanadas"]
    for empanada in lista_empanadas:
      print(str(empanada[2]).rjust(2) + " " + empanada[0].ljust(25) + " " + empanada[1]+ "."*(40-len(empanada[1])) + "$ " + str(empanada[3]).rjust(7))
    print()

  # Imprimir detalle de pedidos de bebidas
  if pedido_cliente["bebidas"] != []:
    print("Bebidas:")
    lista_bebidas = pedido_cliente["bebidas"]
    for bebida in lista_bebidas:
      print(str(bebida[1]).rjust(2) + " " + bebida[0].ljust(25)+ "."*(40-len(bebida[0])) + "$ " + str(bebida[2]).rjust(7))
    print()

  print("Total: $ ",pedido_cliente["total"])

  print()
```

### 8.3 - Solicitud de confirmacion del pedido ###

```python
def confirmar_pedido(pedido_cliente):
  # Aca tendria que mostar todos los datos del pedido
  # del cliente y pedirle la confirmacion
  # Tiene que retornar un booleano

  pedido_cliente_con_totales = detalle_pedido_con_totales(pedido_cliente)

  mostrar_detalle_pedido(pedido_cliente_con_totales)

  respuesta = input("¿Confirmas el pedido (Si/No)? ").upper()

  return respuesta == "SI" or respuesta == "S"
```

## 9 - Bloque Principal del Chatbot ##

```python
def pedido_vacio(pedido_cliente):
  pedido_pizzas = pedido_cliente[1]
  pedido_empanadas = pedido_cliente[2]
  pedido_bebidas = pedido_cliente[3]
  return len(pedido_pizzas) == 0 and len(pedido_empanadas) == 0 and len(pedido_bebidas) == 0

#-------------------------------------------------------------
# Bloque Principal
#-------------------------------------------------------------

# Lista de pedidos recibidos
pedidos_recibidos = []

# Numero de ultimo pedido
nro_pedido = 0
# Condicion de fin del bot que un cliente no pida nada

print('='*20 + ' Inicio del Chatbot ' + '='*20 )
print()

hay_mas_pedidos = True

while hay_mas_pedidos:

  nro_pedido += 1

  print()
  print(f'------------ Tomando Nuevo Pedido - Pedido # {nro_pedido} -----------------------')

  mensaje_bienvenida_cliente()

  pedido_cliente = tomar_pedido(nro_pedido)

  if not pedido_vacio(pedido_cliente):
    pedido_confirmado = confirmar_pedido(pedido_cliente)

    if pedido_confirmado:
      pedidos_recibidos.append(pedido_cliente)

    mensaje_despedida_cliente(pedido_confirmado)
  else:
    hay_mas_pedidos = False

  # Si el ultimo cliente no pidio nada termina el chatbot
  hay_mas_pedidos = not pedido_vacio(pedido_cliente)

print('='*20 + '  Fin del Chatbot   ' + '='*20 )
```

## 10 - Listado de Pedidos Recibidos (verificacion) ##

Codigo para listar los pedidos recibidos. Se usa para verificar la lista de pedidos registrados por el bot.

```python
def tipo_pizza(codigo):
  if codigo == 'G':
    return 'Grande'
  else:
    return 'Chica'

def tipo_empanada(codigo):
  if codigo == 'F':
    return 'Frita'
  else:
    return 'Horno'

def listar_pizzas(pizzas):
  for pizza in pizzas:
    item = buscar_lista_precios_x_codigo(lista_precios_pizzas, pizza[0])
    descripcion = item['descripcion']
    print(' '*3 + descripcion.ljust(25) + ' - Tamaño: ' + tipo_pizza(pizza[1]) + ' - Cantidad: ' + str(pizza[2]))

def listar_empanadas(empanadas):
  for empanada in empanadas:
    item = buscar_lista_precios_x_codigo(lista_precios_empanadas, empanada[0])
    descripcion = item['descripcion']
    print(' '*3 + descripcion.ljust(25) + ' - Tipo: ' + tipo_empanada(empanada[1]) + ' - Cantidad: ' + str(empanada[2]))

def listar_bebidas(bebidas):
  for bebida in bebidas:
    item = buscar_lista_precios_x_codigo(lista_precios_bebidas, bebida[0])
    descripcion = item['descripcion']
    cantidad = str(bebida[1])
    print(' '*3 + descripcion.ljust(25) + ' - Cantidad: ' + cantidad)


def listar_pedidos(pedidos):
  print()
  print('Pedidos recibidos')
  print('='*60)
  for pedido in pedidos:
    if pedido != []:
      linea_pedido = pedido[0]
      pizzas = pedido[1]
      empanadas = pedido[2]
      bebidas = pedido[3]
      print('Pedido # ' + str(linea_pedido))

      if pizzas != []:
        print(" Pizzas:")
        listar_pizzas(pizzas)

      if empanadas != []:
        print(" Empanadas:")
        listar_empanadas(empanadas)

      if bebidas != []:
        print(" Bebidas:")
        listar_bebidas(bebidas)
  print('='*60)

listar_pedidos(pedidos_recibidos)
```
