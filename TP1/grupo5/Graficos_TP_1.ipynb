```python
categorias_productos=['Pizzas', 'Empanadas', 'Bebidas']

total_ventas_pizzas = 0.0
total_ventas_empanadas = 0.0
total_ventas_bebidas = 0.0

for pedido in pedidos_recibidos:
  pizzas = pedido['pizzas']
  for pizza in pizzas:
    total_ventas_pizzas = total_ventas_pizzas + pizza[3]

  empanadas = pedido['empanadas']
  for empanada in empanadas:
    total_ventas_empanadas = total_ventas_empanadas + empanada[3]

  bebidas = pedido['bebidas']
  for bebida in bebidas:
    total_ventas_bebidas = total_ventas_bebidas + bebida[2]

totales_x_ventas = [total_ventas_pizzas, total_ventas_empanadas, total_ventas_bebidas]

totales_x_ventas_en_miles = [0]*len(totales_x_ventas)
for ind in range(len(totales_x_ventas)):
  totales_x_ventas_en_miles[ind] = totales_x_ventas[ind] / 1000.0
```

```python
import matplotlib.pyplot as plt


colores=["#EE6055","#60D394","#AAF683","#FFD97D","#FF9B85"]

def autopct_fun(abs_values):
    gen = iter(abs_values)
    return lambda pct: f"{pct:.2f}% ({next(gen)})"

fig, ax=plt.subplots(figsize=(13,9))
plt.pie(totales_x_ventas,labels=categorias_productos,
        autopct=autopct_fun(totales_x_ventas_en_miles),
        colors=colores,
        shadow=True)
ax.set_title("Ventas por Categoria de Producto (en miles de pesos)", fontsize=15, fontweight="bold")
plt.rcParams.update({'font.size': 12})
ax.legend()
plt.show()
```
