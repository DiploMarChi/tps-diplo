```python
def buscar_lista_precios_x_codigo(lista_precios, codigo):
  for item in lista_precios:
      if item["codigo"] == codigo:
        return item

def codigos_lista_precios(lista_precios):
  codigos = []
  for item in lista_precios:
      codigos.append(item["codigo"])
  return codigos
```

```python
def detalle_pedido_pizzas(pedido_cliente):
  # Arma el detalle de las pizzas
  pizzas=[]
  total_pizzas=0
  for pizza in pedido_cliente[1]:
    item = buscar_lista_precios_x_codigo(lista_precios_pizzas, pizza[0])
    descripcion = item['descripcion']
    precio_pizza_grande = item['grande']
    precio_pizza_chica = item['chica']
    cantidad = pizza[2]
    if pizza[1] == 'G':
      tamaño = 'grande'
      monto = cantidad * precio_pizza_grande
    else:
      tamaño = 'chica'
      monto = cantidad * precio_pizza_chica
    pizzas.append([descripcion, tamaño, cantidad, monto])
    total_pizzas = total_pizzas + monto

  return {'pizzas': pizzas, 'total_pizzas': total_pizzas}

def detalle_pedido_empanadas(pedido_cliente):
  empanadas=[]
  total_empanadas=0
  for empanada in pedido_cliente[2]:
    item = buscar_lista_precios_x_codigo(lista_precios_empanadas, empanada[0])

    descripcion = item['descripcion']
    tipo = empanada[1]
    precio_horno = item['horno']
    precio_frita = item['frita']
    cantidad = empanada[2]
    if tipo == 'F':
      tipo = 'frita'
      monto = cantidad * precio_frita
    else:
      tipo = 'horno'
      monto = cantidad * precio_horno

    empanadas.append([descripcion, tipo, cantidad, monto])
    total_empanadas = total_empanadas + monto

  return {'empanadas': empanadas, 'total_empanadas': total_empanadas}

def detalle_pedido_bebidas(pedido_cliente):
  bebidas=[]
  total_bebidas=0
  for bebida in pedido_cliente[3]:
    item = buscar_lista_precios_x_codigo(lista_precios_bebidas, bebida[0])

    descripcion = item['descripcion']
    precio = item['litro']

    cantidad = bebida[1]
    monto = cantidad * precio

    bebidas.append([descripcion, cantidad, monto])
    total_bebidas = total_bebidas + monto

  return {'bebidas': bebidas, 'total_bebidas': total_bebidas}


def detalle_pedido_con_totales(pedido_cliente):
  # Hacer el calculo de los subtotales de pizzas, empanadas y bebidas
  # y el total del pedido

  # Arma el detalle de las pizzas
  detalle_pizzas = detalle_pedido_pizzas(pedido_cliente)
  pizzas = detalle_pizzas["pizzas"]
  total_pizzas = detalle_pizzas["total_pizzas"]

  # Arma el detalle de las empanadas
  detalle_empanadas = detalle_pedido_empanadas(pedido_cliente)
  empanadas = detalle_empanadas["empanadas"]
  total_empanadas = detalle_empanadas["total_empanadas"]

  # Arma el detalle de las bebidas
  detalle_bebidas = detalle_pedido_bebidas(pedido_cliente)
  bebidas = detalle_bebidas["bebidas"]
  total_bebidas = detalle_bebidas["total_bebidas"]

  total_pedido = total_pizzas + total_empanadas + total_bebidas

  detalle_pedido = {"nro_pedido":pedido_cliente[0],"pizzas":pizzas,"empanadas":empanadas,"bebidas":bebidas,"total":total_pedido}

  return detalle_pedido
```

```python
def generar_pedidos_pizzas():
  pizzas = []
  for n in range(rd.randint(1,3)):
      variedad = rd.randint(1,7)
      random_tamaño = rd.randint(1,2)
      if random_tamaño == 1:
        tamaño = 'G'
      else:
        tamaño = 'C'
      cantidad = rd.randint(1,5)
      pizza = [variedad, tamaño, cantidad]
      pizzas.append(pizza)
  return pizzas

def generar_pedidos_empanadas():
  empanadas = []
  for n in range(rd.randint(1,3)):
      variedad = rd.randint(1,7)
      random_tipo = rd.randint(1,2)
      if random_tipo == 1:
        tipo = 'F'
      else:
        tipo = 'H'
      cantidad = rd.randint(1,12)
      empanada = [variedad, tipo, cantidad]
      empanadas.append(empanada)
  return empanadas

def generar_pedidos_bebidas():
  bebidas = []
  for n in range(rd.randint(1,3)):
      variedad = rd.randint(1,9)
      cantidad = rd.randint(1,5)
      bebida = [variedad, cantidad]
      bebidas.append(bebida)
  return bebidas
```

```python
cantidad_pedidos = 100

lista_pedidos = []
print('[')
for nro_pedido in range(1,cantidad_pedidos+1):
  pizzas = generar_pedidos_pizzas()
  empanadas = generar_pedidos_empanadas()
  bebidas = generar_pedidos_bebidas()
  pedido_cliente = [nro_pedido, pizzas, empanadas, bebidas]
  detalle_pedido = detalle_pedido_con_totales(pedido_cliente)
  if nro_pedido < cantidad_pedidos:
    print(str(detalle_pedido) + ",")
  else:
    print(detalle_pedido)
  lista_pedidos.append(detalle_pedido)
  lista_pedidos.append(detalle_pedido)
print(']')

print()
print()

print(lista_pedidos)


```
